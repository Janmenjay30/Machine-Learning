```python
from nltk.stem import PorterStemmer

ps=PorterStemmer()

```

```python
word="Hi everyone i am janmenjay from IIIT Bhubaneswar  studying Btech, eating chicken"
words=word.split(" ")
for w in words:
    print(w,"-->",ps.stem(w))
```

```python
 ps.stem("sitting")
```

```python
from nltk.stem import RegexpStemmer

reg_stemmer=RegexpStemmer("ing$|s$|ed$",min=4)
```

```python
reg_stemmer.stem('waiting')
reg_stemmer.stem('added to matching dresses')
```
