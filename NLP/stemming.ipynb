```python
from nltk.stem import PorterStemmer

ps=PorterStemmer()

```

```python
word="Hi everyone i am janmenjay from IIIT Bhubaneswar  studying Btech, eating chicken"
words=word.split(" ")
for w in words:
    print(w,"-->",ps.stem(w))
```

```python
from nltk.stem import RegexpStemmer

reg_stemmer=RegexpStemmer("ing$|s$|ed$",min=4)
```

```python
reg_stemmer.stem('waiting')
reg_stemmer.stem('added to matching dresses')
```

```python
import nltk
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer=WordNetLemmatizer()

```

```python
word=["cat","cats","cat's","cats"]

for w in word:
    print(lemmatizer.lemmatize(w))
```

```python
word=["eating","eats","ate"]

for w in word:
    print(w,":",lemmatizer.lemmatize(w,pos='v'))
```
