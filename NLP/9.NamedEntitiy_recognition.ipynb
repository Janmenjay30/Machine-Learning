# Named Entity Recognition (NER)

Named Entity Recognition is a process where we identify and classify named entities in text into specific categories like names of people, organizations, locations, expressions of times, quantities, monetary values, percentages, etc.

```python
import nltk
from nltk import word_tokenize, pos_tag, ne_chunk

# Download necessary NLTK data
nltk.download('maxent_ne_chunker_tab')
nltk.download('words')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
sentence = "The Eiffel Tower was built by Gustave Eiffel and is located in Paris, France."

paragraph = """
Elon Musk, the CEO of SpaceX and Tesla, announced that the companies will be collaborating with NASA for the upcoming Mars mission scheduled for 2030. 
The launch will take place from the Kennedy Space Center in Florida. 
Meanwhile, Google and Microsoft are also investing heavily in AI research in Silicon Valley.
"""
```

```python
word_tokenize(sentence)
```

```python
words=word_tokenize(sentence)
```

```python
for w in words:
    print(f"{w}")
```

```python
tagged_elements=nltk.pos_tag(words)
```

```python
nltk.download('words')
```

```python
nltk.ne_chunk(tagged_elements)
```
